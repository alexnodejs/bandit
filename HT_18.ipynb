```python
from diffusers import AutoPipelineForText2Image
import torch

pipeline = AutoPipelineForText2Image.from_pretrained(
    "stable-diffusion-v1-5/stable-diffusion-v1-5"
).to("cpu")
generator = torch.Generator("cpu").manual_seed(32)
image = pipeline("A roof cat living in a big city (Tokyo) spends its days basking in the sun atop skyscrapers and its nights leaping gracefully between rooftops under the glow of neon lights. 4k", generator=generator).images[0]
image
```
