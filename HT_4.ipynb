# Завдання 3
Очистіть дані від викидів. Виведіть очищений датафрейм

```python
df2= df.copy()
num_columns = df2.select_dtypes(include="number").columns

num = df2[num_columns]
q1 = num.quantile(q=0.25)
q3 = num.quantile(q=0.75)

iqr = q3 - q1

lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

mask = (num >= lower) & (num <= upper)
mask = mask.all(axis=1)

df_clean = df[mask]
df_clean
```
