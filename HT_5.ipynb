```python
df = pd.DataFrame(df)

# Numerical features: boxplots and histograms
plt.figure(figsize=(10, 8))

# Boxplot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(df['age'])
plt.title('Boxplot of age')

# Histogram for 'age'
plt.subplot(1, 2, 2)
sns.histplot(df['age'], bins=12, kde=True)
plt.title('Histogram of age')
plt.xlabel('age')

plt.tight_layout()
plt.show()
```

```python
# Categorical features: group by 'sex' and analyze 'embarked'
plt.figure(figsize=(8, 8))
sns.boxplot(x='sex', y='embarked', data=df)
plt.title('Boxplot of sex by embarked')
plt.xlabel('Passenger sex')
plt.ylabel('embarked')
plt.show()
```
