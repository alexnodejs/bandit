```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Reinforcment

# Combine relevant columns to create unique "arms"
data['arm'] = (
    data['Brand'] + "_" +
    data['Channel'] + "_" +
    data['Category'] + "_" +
    data['Spot_language']
)

# We can drop this because we are using OneHotEncoder later , however we can experiment
# # Encode categorical variables (state features)
# encoder = LabelEncoder()
# data['state'] = encoder.fit_transform(data['Category'] + "_" + data['Spot_language'])

data

```

```python
# Кодування категорій за допомогою get_dummies
encoded_data = pd.get_dummies(data[['Brand', 'Channel', 'Category', 'Spot_language']], drop_first=True)

# Об'єднання з основним DataFrame
data_with_encoded = pd.concat([data, encoded_data], axis=1)

# Видаляємо зайві стовпці, якщо необхідно
data_with_encoded.head()
```

```python
# Ініціалізація параметрів
arms = data['arm'].unique()
num_arms = len(arms)
weights = np.ones(num_arms)  # Початкові ваги однакові
gamma = 0.1  # Рівень дослідження

# Мапа для індексів рук
arm_to_index = {arm: i for i, arm in enumerate(arms)}

# Функція для розрахунку ймовірностей
def calculate_probabilities(weights, gamma):
    total_weight = np.sum(weights)
    probabilities = (1 - gamma) * (weights / total_weight) + (gamma / num_arms)
    return probabilities

# Моделювання винагород для різних користувачів
class User:
    def __init__(self, preferences):
        self.preferences = preferences

    def reward(self, arm):
        return self.preferences.get(arm, 0)

# Приклад користувачів
# user_preferences = {
#     'Brand_ZEE5_Channel_&pictures_Category_Movies_Spot_language_English': 0.8,
#     'Brand_Sony_Channel_&tv_Category_Sports_Spot_language_Hindi': 0.6,
# }
np.random.seed(42)  # Фіксуємо seed для повторюваності результатів
user_preferences = {arm: np.random.uniform(0, 1) for arm in data['arm'].unique()}
user = User(user_preferences)



# Навчання без класифікатора
T = 1000  # Кількість ітерацій
reward_history = []

for t in range(T):
    probabilities = calculate_probabilities(weights, gamma)
    selected_arm = np.random.choice(arms, p=probabilities)
    arm_index = arm_to_index[selected_arm]

    # Отримання винагороди
    reward = user.reward(selected_arm)
    estimated_reward = reward / probabilities[arm_index] if probabilities[arm_index] > 0 else 0

    # Оновлення ваг
    weights[arm_index] *= np.exp((gamma * estimated_reward) / num_arms)
    reward_history.append(reward)

    # Друк результатів для кожної ітерації
    print(f"Iteration {t+1}: Selected arm = {selected_arm}, Reward = {reward}, Estimated reward = {estimated_reward:.2f}")

# Сумарна винагорода
print(f"Total reward after {T} rounds: {sum(reward_history)}")
print("\nTotal rewards for each arm:")
# for arm, total_reward in reward_history.items():
#     print(f"{arm}: {total_reward}")
```

```python
import matplotlib.pyplot as plt

def plot_rewards(reward_history):
    cumulative_rewards = np.cumsum(reward_history)
    plt.plot(cumulative_rewards)
    plt.title('Cumulative Rewards Over Time')
    plt.xlabel('Iteration')
    plt.ylabel('Cumulative Reward')
    plt.grid(True)
    plt.show()

plot_rewards(reward_history)
```
